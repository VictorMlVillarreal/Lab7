```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
x = df.N_HABITACIONES_MEDIO.plot.hist(edgecolor='black')
x.set_title("Gráfico de barras según el número de habitaciones")
x.set_xlabel("Cantidad de habitaciones")
x.set_ylabel("Frecuencia")
plt.show()
```

```python
x = df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1, edgecolor='black')
plt.title("Gráfico de dispersión por índice de criminalidad")
plt.xlabel("Valor")
plt.ylabel("Índice de crimen")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
x = valor_por_ciudad.head(10).plot.barh(edgecolor='black')
plt.title("Gráfico de barras según ciudad")
plt.xlabel("Cantidad")
plt.ylabel("Ciudad")
plt.show()
```
